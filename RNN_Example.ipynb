```python
#step 1
#import neccesary libraries
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense # Changed simpleRNN to SimpleRNN
import tensorflow as tf
```

```python
#step 2
#generate training data: sequence like[1,2,3] ->4
X=[]
y=[]
for i in range(1,10):
  X.append([i, i+1, i+2])
  y.append([i+3])


X = np.array(X).reshape((len(X),3,1))
Y = np.array(y)
```

```python
# Convert X and y to TensorFlow Datasets
X = tf.data.Dataset.from_tensor_slices(X)
y = tf.data.Dataset.from_tensor_slices(y)

# Create training Dataset.
train_dataset = tf.data.Dataset.zip((X, y)).batch(1) # Batch size of 1

model = Sequential()
model.add(SimpleRNN(10, activation='relu', input_shape=(3,1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Use train_dataset for fitting the model.
model.fit(train_dataset, epochs=300,verbose=1)
```

```python
test_input = np.array([[10, 11, 12]]).reshape((1, 3, 1))
predicted_output = model.predict(test_input,verbose=0)
print(f"Input Sequence:[10,11,12]")
print("Predicted next Number:{round(predicted_output[0][0]),2)}")
```
